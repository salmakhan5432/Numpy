#### Creating Numpy Arrays

```python
# np.array
import numpy as np
a = np.array([1,2,3])
a
```

```python
# 2D and 3D
b = np.array([[1,2,3],[3,4,5]])
b
```

```python
c = np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print(c)
```

```python
# dtype
np.array([1,2,3],dtype=float)
```

```python
# np.arange
np.arange(1,11,2)
```

```python
# with reshape
np.arange(16).reshape(2,2,2,2)
```

```python
# np.ones and np.zeros
o = np.ones((3,4))
print(o)
```

```python
z = np.zeros((3,4))
print(z)
```

```python
# np.random
np.random.random((3,4))
```

```python
# np.linspace
np.linspace(-10,10,10,dtype=int)
```

```python
# np.identity
i = np.identity(3)
i
```
