
### Array Function

```python

import numpy as np
```

```python
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
 # max/min/sum/prod
# 0 -> column and 1 -> rows
print(np.max(a1, axis= 1))
print(np.min(a1))
print(np.sum(a1))
print(np.prod(a1))
```

```python
# mean/median/std/var
np.median(a1, axis= 1)
```

```python
# Trigonometric function
np.sin(a1)
```

```python
# Dot product
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
np.dot(a2,a3)
```

```python
# log and exponent
(np.log(a1))
```

```python
#round/floor/ceil
np.round(np.random.random((2,3))* 100)    # Round: It returns to the nearest integer
```

```python
np.floor(np.random.random((2,3)) * 100)
```

### Stacking

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
np.hstack((a4,a5))
```

```python
# vertical stacking
np.vstack((a4,a5))  # shape must be same to do stacking
```

### Splitting

```python
# horizontal split
np.hsplit(a4,2)
```

```python
# vertical split
print(np.vsplit(a4,3))
```
