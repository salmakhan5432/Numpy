### Advance Indexing

```python
import numpy as np
```

```python
a = np.arange(24).reshape(6,4)
print(a)
```

```python
# Fancy Indexing
print(a[[0,2,5]])
```

```python
print(a[:,[2,3]])
```

```python
# Boolean Indexing
a = np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# Find all number greater than 50
a[a > 50]
```

```python
# Find out even numbers
a[a % 2 == 0]
```

```python
# Find all numbers greater than 50 and are even
a[(a > 50) & (a % 2 == 0)]
```

```python
# Find all number that are not divisible by 7
a[~(a % 7 == 0)]
```

### Broadcasting
The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
 The smaller array is "broadcast" across the larger array so that they have compatible shapes

```python
# Same shape
a = np.arange(6).reshape(3,2)
b = np.arange(6,12).reshape(3,2)
print(a)
print(b)
print(a + b)
```

```python
# Diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a + b)
```

### Broadcasting Rules
#### 1. Make the two arrays have same number of dimensions.
      If the numbers of dimensions of the two arrays are different, add new dimensions with size 1 to the head of the array with the smaller dimension.

#### 2. Make each dimension of two arrays the same size.
      If the size of each dimension of the two arrays do not match, dimension with size 1 are stretched to the size of the other array.
      If there is a dimension whose size is not 1 either of the two arrays, it cannot be broadcasted, and an error is raised.

```python
# More Examples
a = np.arange(12).reshape(4,3)
b = np.arange(3)
print(a + b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)
print(a + b)
```

```python
a = np.array([1]) # Shape ---> [1,1]
b = np.arange(4).reshape(2,2)
print(a + b)
```

###  Sigmoid Function

```python
# Sigmoid
def sigmoid(array):
    return 1/(1 + np.exp(-(array)))
a = np.arange(10)
sigmoid(a)
```

### Mean Squared Error (MSE)

```python
# Mean square error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
def mse(actual,predicted):
    return np.mean((actual - predicted)**2)
print(mse(actual,predicted))
```

### Working with Missing Values in NumPy
#####   In real-world datasets, missing values are extremely common.
#####   NumPy represents missing or undefined numerical values using np.nan (Not a Number).

```python
# Working with missing values ---> np.nan
a = np.array([1,2,3,4,5,np.nan,6,7])
a
```

```python
np.isnan(a)
```

```python
a[np.isnan(a)]
```

```python
a[~np.isnan(a)]
```
